# Linear Algebra

# Vectors & Matrices

The following notation is used.

\begin{align}
a & \in \mathbb R \qquad & \text{a scalar} \\
\mathbf u & \in \mathbb R^N   \qquad & \text{a vector} \\
\mathbf U & \in \mathbb R^{N\times M}  \qquad & \text{a matrix}
\end{align}

Alternatively, vectors may be written as $\vec v$.

Vectors and matrices are one and two dimensinal sequences of vectors.

$$
\mathbf v = 
\begin{matrix}
1 \\
0
\end{matrix}
\qquad \qquad
\mathbf U = 
\begin{matrix}
1 & 2\\
3 & 4
\end{matrix}
$$

Matrices can be transposed. Transposing a column-vector yields a row-vector.

$$
\mathbf v^T = 
\begin{matrix}
1 & 0
\end{matrix}
$$

Matrix can be inversed.
A special matrix is the identity matrix.

$$
\mathbf {AA}^{-1} = I =
\begin{matrix}
1 & 0\\
0 & 1
\end{matrix}
$$

$$ (\mathbf {AB})^{-1} = \mathbf B^{-1} \mathbf A^{-1}$$


# Matrix Operations

## Inner product

Inner product (dot product): $\mathbf u^T \mathbf v$ or $\vec u \cdot \vec v$.

- The projection of $u$ onto $v$ and...
- The product of the lengths and the angle: $\|u\| \|v\| \cos \theta$

## Outer product

$\mathbf{AB} = C$

$\mathbf{AB}^T = \mathbf B^T \mathbf A^T$

## Cross product

The cross product denotes the area between two vectors: $\mathbf u \times \mathbf v$

$$
\| \mathbf u \times \mathbf v \| = \| \mathbf u \| \| \mathbf v \| \sin \theta
$$

Direction - right hand rule (by convention).

\begin{aligned}
\mathbf x \times \mathbf x &= 0 \\
\mathbf x \times \mathbf y &= \mathbf z \\
\mathbf y \times \mathbf x &= - \mathbf z \\
\mathbf y \times \mathbf z &= \mathbf x \\
&\vdots
\end{aligned}


Dual matrix

$$
\mathbf u \times \mathbf b = \mathbf U^* \, \mathbf b
$$
